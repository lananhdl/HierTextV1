draw_mask: tạo mask nhị phân từ tọa độ vertices (đa giác bao quanh từ/dòng/đoạn).

stack_word_masks: cộng dồn các mask và biến thành nhị phân (1 nếu có từ nào nằm trong vùng).

```python
# 5. Hàm xử lý mask, annotation

def draw_mask(vertices, w, h):
    mask = np.zeros((h, w, 3), dtype=np.float32)
    mask = cv2.fillPoly(mask, [vertices], [1.] * 3)[:, :, 0]
    return mask

def stack_word_masks(mask, word_mask):
    return (mask + word_mask > 0.).astype(np.float32)

def load_img(image_id):
    img_path = os.path.join(IMAGE_SUBSET_DIR_PATH, f'{image_id}.jpg')

    if not os.path.exists(img_path):
        raise FileNotFoundError(f"Không tìm thấy ảnh: {img_path}")

    image = cv2.imread(img_path)
    if image is None:
        raise ValueError(f"Không thể đọc ảnh từ: {img_path}. Có thể file bị lỗi hoặc không phải ảnh hợp lệ.")

    return image
```

Hàm xử lý annotation để tách mask và văn bản

Trích xuất mask, text, và trọng số (weights) từ annotation ở 3 cấp: word, line, paragraph.

Trọng số = 1 nếu legible (đọc được), ngược lại là 0.

Tạo ra các dictionary word_dict, line_dict, paragraph_dict để dùng cho trực quan hóa hoặc đánh giá sau.

```python
import cv2  # Thêm thư viện OpenCV để vẽ đường viền
import numpy as np  # Thư viện NumPy để xử lý mảng

def parse_annotation_dict(ann):
    # Định nghĩa hàm để phân tích từ điển chú thích và trích xuất dữ liệu liên quan đến hình ảnh và văn bản
    image_id = ann['image_id']  # Lấy ID của hình ảnh từ từ điển chú thích
    print(f'Processing image {image_id}')  # In thông báo đang xử lý hình ảnh với ID tương ứng
    image = load_img(image_id)  # Tải hình ảnh bằng ID (giả sử hàm load_img được định nghĩa ở nơi khác)
    w = ann['image_width']  # Lấy chiều rộng của hình ảnh từ từ điển chú thích
    h = ann['image_height']  # Lấy chiều cao của hình ảnh từ từ điển chú thích

    gt_word_masks = []  # Khởi tạo danh sách rỗng để lưu các mask nhị phân cấp độ từ
    gt_word_weights = []  # Khởi tạo danh sách rỗng để lưu trọng số của từ (1.0 nếu đọc được, 0.0 nếu không đọc được)
    gt_word_texts = []  # Khởi tạo danh sách rỗng để lưu nội dung văn bản của các từ

    gt_line_masks = []  # Khởi tạo danh sách rỗng để lưu các mask nhị phân cấp độ dòng
    gt_line_weights = []  # Khởi tạo danh sách rỗng để lưu trọng số của dòng (1.0 nếu đọc được, 0.0 nếu không đọc được)
    gt_line_texts = []  # Khởi tạo danh sách rỗng để lưu nội dung văn bản của các dòng

    gt_paragraph_masks = []  # Khởi tạo danh sách rỗng để lưu các mask nhị phân cấp độ đoạn văn
    gt_paragraph_weights = []  # Khởi tạo danh sách rỗng để lưu trọng số của đoạn văn (1.0 nếu đọc được, 0.0 nếu không đọc được)

    # Sao chép hình ảnh để vẽ đường viền, tránh thay đổi hình ảnh gốc
    annotated_image = image.copy()

    for paragraph in ann['paragraphs']:  # Duyệt qua từng đoạn văn trong chú thích
        gt_paragraph_mask = np.zeros((h, w), dtype=np.float32)  # Tạo mask rỗng cho đoạn văn với kích thước hình ảnh
        # Lấy tọa độ đỉnh của đoạn văn và vẽ đường viền màu xanh
        if 'vertices' in paragraph:  # Kiểm tra xem đoạn văn có tọa độ đỉnh không
            paragraph_vertices = np.array(paragraph['vertices'], dtype=np.int32)  # Chuyển tọa độ đỉnh thành mảng NumPy
            paragraph_vertices = paragraph_vertices.reshape((-1, 1, 2))  # Định dạng lại thành dạng phù hợp với cv2.polylines
            cv2.polylines(annotated_image, [paragraph_vertices], isClosed=True, color=(0, 255, 0), thickness=2)  # Vẽ đường viền màu xanh (BGR: 0, 255, 0) cho đoạn văn
        for line in paragraph['lines']:  # Duyệt qua từng dòng trong đoạn văn hiện tại
            gt_line_mask = np.zeros((h, w), dtype=np.float32)  # Tạo mask rỗng cho dòng với kích thước hình ảnh
            # Lấy tọa độ đỉnh của dòng và vẽ đường viền màu vàng
            if 'vertices' in line:  # Kiểm tra xem dòng có tọa độ đỉnh không
                line_vertices = np.array(line['vertices'], dtype=np.int32)  # Chuyển tọa độ đỉnh thành mảng NumPy
                line_vertices = line_vertices.reshape((-1, 1, 2))  # Định dạng lại thành dạng phù hợp với cv2.polylines
                cv2.polylines(annotated_image, [line_vertices], isClosed=True, color=(0, 255, 255), thickness=2)  # Vẽ đường viền màu vàng (BGR: 0, 255, 255) cho dòng
            for word in line['words']:  # Duyệt qua từng từ trong dòng hiện tại
                gt_word_weights.append(1.0 if word['legible'] else 0.0)  # Thêm trọng số (1.0 nếu từ đọc được, 0.0 nếu không)
                gt_word_texts.append(word['text'])  # Thêm nội dung văn bản của từ
                vertices = np.array(word['vertices'])  # Chuyển tọa độ đỉnh của từ thành mảng NumPy
                gt_word_mask = draw_mask(vertices, w, h)  # Tạo mask nhị phân cho từ dựa trên tọa độ đỉnh (giả sử draw_mask được định nghĩa)
                gt_word_masks.append(gt_word_mask)  # Thêm mask của từ vào danh sách
                gt_line_mask = stack_word_masks(gt_line_mask, gt_word_mask)  # Kết hợp mask của từ vào mask của dòng (giả sử stack_word_masks được định nghĩa)
                gt_paragraph_mask = stack_word_masks(gt_paragraph_mask, gt_word_mask)  # Kết hợp mask của từ vào mask của đoạn văn
            gt_line_masks.append(gt_line_mask)  # Thêm mask của dòng vào danh sách
            gt_line_weights.append(1.0 if line['legible'] else 0.0)  # Thêm trọng số (1.0 nếu dòng đọc được, 0.0 nếu không)
            gt_line_texts.append(line['text'])  # Thêm nội dung văn bản của dòng
        if not paragraph['legible']:  # Kiểm tra nếu đoạn văn không đọc được
            gt_paragraph_mask = draw_mask(np.array(paragraph['vertices']), w, h)  # Tạo mask cho đoạn văn dựa trên tọa độ đỉnh
        gt_paragraph_masks.append(gt_paragraph_mask)  # Thêm mask của đoạn văn vào danh sách
        gt_paragraph_weights.append(1.0 if paragraph['legible'] else 0.0)  # Thêm trọng số (1.0 nếu đoạn văn đọc được, 0.0 nếu không)

    num_gt_words = len(gt_word_masks)  # Đếm số lượng mask của từ
    word_dict = {  # Tạo từ điển chứa dữ liệu liên quan đến từ
        'gt_weights': (np.array(gt_word_weights) if num_gt_words else np.zeros((0,), np.float32)),  # Chuyển trọng số từ thành mảng NumPy hoặc mảng rỗng nếu không có
        'gt_masks': (np.stack(gt_word_masks, -1) if num_gt_words else np.zeros(((h + 1) // 2, (w + 1) // 2, 0), np.float32)),  # Gộp các mask từ hoặc tạo mask rỗng với kích thước giảm một nửa
        'gt_texts': (np.array(gt_word_texts) if num_gt_words else np.zeros((0,), str)),  # Chuyển văn bản từ thành mảng NumPy hoặc mảng rỗng nếu không có
    }

    num_gt_lines = len(gt_line_masks)  # Đếm số lượng mask của dòng
    line_dict = {  # Tạo từ điển chứa dữ liệu liên quan đến dòng
        'gt_weights': (np.array(gt_line_weights) if num_gt_lines else np.zeros((0,), np.float32)),  # Chuyển trọng số dòng thành mảng NumPy hoặc mảng rỗng nếu không có
        'gt_masks': (np.stack(gt_line_masks, -1) if num_gt_lines else np.zeros(((h + 1) // 2, (w + 1) // 2, 0), np.float32)),  # Gộp các mask dòng hoặc tạo mask rỗng với kích thước giảm một nửa
        'gt_texts': (np.array(gt_line_texts) if num_gt_lines else np.zeros((0,), str)),  # Chuyển văn bản dòng thành mảng NumPy hoặc mảng rỗng nếu không có
    }

    num_gt_paragraphs = len(gt_paragraph_masks)  # Đếm số lượng mask của đoạn văn
    paragraph_dict = {  # Tạo từ điển chứa dữ liệu liên quan đến đoạn văn
        'gt_weights': (np.array(gt_paragraph_weights) if num_gt_paragraphs else np.zeros((0,), np.float32)),  # Chuyển trọng số đoạn văn thành mảng NumPy hoặc mảng rỗng nếu không có
        'gt_masks': (np.stack(gt_paragraph_masks, -1) if num_gt_paragraphs else np.zeros(((h + 1) // 2, (w + 1) // 2, 0), np.float32)),  # Gộp các mask đoạn văn hoặc tạo mask rỗng với kích thước giảm một nửa
    }

    return annotated_image, word_dict, line_dict, paragraph_dict  # Trả về hình ảnh đã được vẽ đường viền và các từ điển chứa dữ liệu từ, dòng, đoạn văn
```

```python
# 6. Hàm trực quan hóa

def visualize(text_dict):
  h, w, n_mask = text_dict['gt_masks'].shape
  if not n_mask:
    return np.zeros((h, w, 3), np.float32)
  palette = np.random.uniform(0.0, 1., (n_mask, 3))
  colored = np.reshape(np.matmul(
        np.reshape(text_dict['gt_masks'], (-1, n_mask)), palette), (h, w, 3))
  dont_care_mask = (np.reshape(np.matmul(
        np.reshape(text_dict['gt_masks'], (-1, n_mask)),
        np.reshape(1.- text_dict['gt_weights'], (-1, 1))), (h, w, 1)) > 0).astype(np.float32)

  filtered = np.clip(dont_care_mask * 1. + (1. - dont_care_mask) * colored, 0., 1.)

  plt.figure(figsize=(FIG_SIZE, FIG_SIZE))
  plt.imshow(filtered)
  plt.show()
```

```python
# 7. Gọi trực quan hóa cho một ảnh mẫu random
import random

SAMPLE_ID = random.randint(0, len(anns) - 1)

image, word_dict, line_dict, para_dict = parse_annotation_dict(anns[SAMPLE_ID])

# Hiển thị ảnh gốc
plt.figure(figsize=(FIG_SIZE, FIG_SIZE))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis("off")
plt.show()

# Hiển thị word-level annotations
print("Word annotation:")
visualize(word_dict)

# Hiển thị line-level
print("Line grouping:")
visualize(line_dict)

# Hiển thị paragraph-level
print("Paragraph grouping:")
visualize(para_dict)
import json
import os

# Tạo thư mục nếu chưa tồn tại
os.makedirs("/content/hiertext/output", exist_ok=True)

# Lấy annotation gốc của SAMPLE_ID
sample_ann = anns[SAMPLE_ID]

# Đóng gói thành định dạng predictions
predictions = {
    "annotations": [sample_ann]
}

# Ghi ra file predictions.json
with open("/content/hiertext/output/predictions.json", "w", encoding="utf-8") as f:
    json.dump(predictions, f, indent=2, ensure_ascii=False)

print(f"Đã ghi predictions.json cho image_id = {sample_ann['image_id']}")
```
